```python
from pyecharts.charts import Funnel
from pyecharts.globals import ThemeType
from pyecharts.faker import Faker
import pyecharts.options as opts
from random import randint

# 漏斗图适用于业务流程比较规范、周期长、环节多的流程分析，通过漏斗各环节业务数据的比较，能够直观地发现和说明问题所在
funnel = Funnel(init_opts=opts.InitOpts(theme=ThemeType.DARK))
funnel.add('',[list(z) for z in zip(Faker.choose(),Faker.values())])
funnel.render_notebook()
```

```python
funnel = Funnel(init_opts=opts.InitOpts(theme=ThemeType.DARK))
funnel.add('',[list(z) for z in zip(Faker.choose(),Faker.values())],
           label_opts=opts.LabelOpts(position='inside',formatter='{b}:{c}',color='black',),
           sort_='ascending')
funnel.render_notebook()
```
