```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'SimHei'

x = ['a','b','c','d']
y = [1,2,3,4]
# 方差
yerr = [0.2,0.4,0.5,0.6] 


# 设置y轴的范围
m = max(zip(y,yerr))
plt.ylim([0,(m[0]+m[1])*2])

plt.bar(x,y,yerr=yerr)
```
