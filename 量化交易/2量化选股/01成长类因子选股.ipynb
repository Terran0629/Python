营业收入同比增长率选股

同比增长率是相对于上一年度的同一期的数据

```python
from jqdatasdk import *

# indicator为财务指标数据库
q = query(indicator).filter(
    indicator.inc_revenue_year_on_year > 800
)

df = get_fundamentals(q,'2023-01-01')
df.head()
```
