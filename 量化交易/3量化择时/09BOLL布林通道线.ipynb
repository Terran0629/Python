BOLL布林通道线

组成：上轨,中轨,下轨

```python
'''
中轨线：收盘价的20日简单移动平均线(默认20天)
上轨线：中轨线+2倍的收盘价的20日估算标准差(默认2倍)
下轨线：中轨线-2倍的收盘价的20日估算标准差(默认2倍)
指标原理：布林通道内可以覆盖约95%的价格(在上下2个标准差的情况下)

BOLL指标用法
股价上升穿越布林通道线上轨时,回档(降价)概率大
股价下跌穿越布林通道线下轨时,反弹概率大

注意
指标是跟随价格变化的,当下的走势会有各种变化,但是过去的走势是不变的,
所以使用布林指标去复盘很好,但是进行当下的交易要慎重
要关注布林带所统计的数据,数据越稳定,统计的价值越大
'''
```
