市净率选股

市净率=每股股价/每股净资产

```python
from jqdatasdk import *

# valuation为市值表数据库
q = query(valuation).filter(
    valuation.pb_ratio < 1
)

df = get_fundamentals(q,'2023-01-01')
df.head()
```
