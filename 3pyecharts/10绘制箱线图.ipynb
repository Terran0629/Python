```python
from pyecharts.charts import Boxplot
from pyecharts.globals import ThemeType
from pyecharts.faker import Faker
import pyecharts.options as opts
from random import randint

a = [[randint(0,100) for _ in range(100)] for _ in range(4)]
# len(a)
box = Boxplot(init_opts=opts.InitOpts(theme=ThemeType.DARK))
box.add_xaxis([f'exp{i}' for i in range(1,5)])
box.add_yaxis('',box.prepare_data(a))
box.render_notebook()
```

```python
box = Boxplot(init_opts=opts.InitOpts(theme=ThemeType.DARK))
box.add_xaxis([f'exp{i}' for i in range(1,5)])
box.add_yaxis('A',box.prepare_data(a))
box.add_yaxis('B',box.prepare_data(a))
box.render_notebook()
```
