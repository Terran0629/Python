```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(context='paper',rc={'lines.linewidth':12})
plt.plot([i for i in range(10)],[i for i in range(10,20)])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'SimHei'
sns.set_context(context='poster')
plt.plot([i for i in range(10)],[i for i in range(10,20)],label='实例')

plt.legend()
```
