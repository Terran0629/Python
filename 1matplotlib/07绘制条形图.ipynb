```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'SimHei'

x = [2020,2025,2030,2035]
y = [10000,50000,35000,60000]

plt.xlabel('年份')
plt.ylabel('销量')

_x = [f'{i}年' for i in x]
_y = [f'{i}元' for i in y]

plt.xticks(x,_x)
plt.yticks(y,_y)

plt.bar(x,y,width=3)
```

```python
import matplotlib.pyplot as plt

plt.rcParams['axes.unicode_minus'] = False    # 用来正常显示负号

x = range(5)
list(x)
y = [5,-9,16,-36,25]

bars = plt.bar(x,y,color='lightblue')

for b,height in zip(bars,y):
    if height<0:
        b.set(color='lightgreen')
```
