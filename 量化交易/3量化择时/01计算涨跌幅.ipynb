```python
import jqdatasdk as jq
import pandas as pd
from jqdatasdk import *
jq.auth('15661317439','Ax020629.')
```

```python
# 获取日K数据
df = get_price('000001.XSHE','2023-01-01','2023-04-01')
df.head()
```

涨跌幅=(当期收盘价−前期收盘价)/前期收盘价

```python
# 日K涨跌幅

df['日K'] = (df['close'] - df['close'].shift(1))/df['close'].shift(1)
df.head()
```

```python
# 周K涨跌幅

df_week = pd.DataFrame()
'''
将数据转换为制定周期:开盘价(周期第1天)、收盘价(周期最后1天)、
最高价(周期内最高价)、最低价(周期内最低价)
'''
df_week['open'] = df['open'].resample('W').first()
df_week['close'] = df['close'].resample('W').last()
df_week['high'] = df['high'].resample('W').max()
df_week['low'] = df['low'].resample('W').min()


df_week['周K'] = (df_week['close'] - df_week['close'].shift(1))/df_week['close'].shift(1)
df_week.head()
```
