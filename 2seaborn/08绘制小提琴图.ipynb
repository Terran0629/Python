```python
import matplotlib.pyplot as plt
import seaborn as sns


data = sns.load_dataset('tips')  # csv

plt.subplot(311)
sns.violinplot(x='day',y='total_bill',data=data)
plt.subplot(312)
sns.violinplot(x='day',y='total_bill',data=data,hue='time') # huex
plt.subplot(313)
sns.violinplot(x='day',y='total_bill',data=data,hue='sex') # huex
```

```python
plt.subplot(121)
sns.violinplot(x='day',y='total_bill',data=data,hue='sex',split=True)
plt.subplot(122)
sns.violinplot(x='day',y='total_bill',data=data,hue='sex',split=True,inner=None)
```
