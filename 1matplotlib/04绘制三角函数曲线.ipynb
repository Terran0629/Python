```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'SimHei'
x = np.linspace(0,10)
# print(x)
_sin = np.sin(x)
_cos = np.cos(x)

plt.rcParams['axes.unicode_minus'] = False    # 用来正常显示负号

plt.plot(x,_sin,label='sin曲线')
plt.plot(x,_cos,label='cos曲线')
plt.legend()
```
