```python
from pyecharts.charts import HeatMap
from pyecharts.globals import ThemeType
from pyecharts.faker import Faker
import pyecharts.options as opts
from random import randint

value = [[i,j,randint(-20,30)] for i in range(24) for j in range(7)]
map = HeatMap(init_opts=opts.InitOpts(theme=ThemeType.DARK))
map.add_xaxis(Faker.clock)
map.add_yaxis('',Faker.week,value)
map.render_notebook()
```
