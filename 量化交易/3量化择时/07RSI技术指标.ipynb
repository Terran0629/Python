RSI技术指标(相对强弱指标)

组成:RSI线,70轴,50轴,30轴

取值范围:0~100

```python
'''
计算公式
RSI = A/(A+B) *100

其中,A=N日内收盘价涨幅之和,B=N日内收盘价跌幅之和的绝对值

例如,获取最近15天K线收盘价,每一根K线收盘价减去上一根K线收盘价,得到14个数值,
则A=近14个数字中正数之和;B=近14个数字中负数之和的绝对值。

根据计算公式,RSI实质上描述的是: 上涨的总波动 / 总波动

注意
RSI的计算过程中,做了加权移动平均,所以距离越近的K线的涨跌幅对RSI的数值的影响越大!

对于RSI指标来说,有价值的区域就是超买区和超卖区
对于起伏不大的稳定市场,RSI超过70轴(RSI > 70),进入超买区;超过30轴(RSI < 30),进入超卖区
30 <RSI <70的中间区域的参考价值不大
由于RSI计算周期的随意性(周期N的设定根据个人偏好)、中间区域的无参考价值,所以RSI指标最好结合其他技术指标使用才更合理

注意
超买：买进股票的人数超过了一定比例。那么,这时候应该反向卖出股票
超卖：代表卖方卖股票过了头,卖股票的人数超过一定比例时，反而应该买进股票
'''
```
