```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np


data = sns.load_dataset('tips')
g = sns.FacetGrid(data=data,row='time',col='sex',hue='smoker'
                  ,palette={'g','b'},hue_kws={'marker':['^','x']})
g.map(sns.scatterplot,'total_bill','tip')
g.add_legend()
```
