```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'SimHei'
x = np.linspace(0,10,100)
# print(x)
sin = np.sin(x)
cos = np.cos(x)

plt.rcParams['axes.unicode_minus'] = False    # 用来正常显示负号

plt.subplot(2,1,1)
plt.plot(x,sin)
plt.subplot(2,1,2)
plt.plot(x,cos)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'SimHei'
x = np.linspace(0,10,100)
# print(x)
sin = np.sin(x)
cos = np.cos(x)
tan = np.tan(x)

plt.rcParams['axes.unicode_minus'] = False    # 用来正常显示负号

fig,ax = plt.subplots(2,2)
ax[0,0].plot(x,sin)
ax[0,1].plot(x,cos)
ax[1,1].plot(x,tan)
```
