```python
import jqdatasdk as jq
import pandas as pd
```

```python
jq.auth('15661317439','Ax020629.')
```

```python
df = jq.get_price("000001.XSHE",count=20,end_date='2023-11-10',frequency="daily")
df
```

```python
# 添加周信息(周一为0)
df['weekday'] = df.index.weekday
df.head(10)
```

```python
# 转换为周数据
df1 = pd.DataFrame()
# 开盘价（当周第一天）
df1['open'] = df.open.resample('W').first()
# 收盘价（当周最后一天）
df1['close'] = df.close.resample('W').last()
# 最高价（当周）
df1['high'] = df.high.resample('W').max()
# 最低价（当周）
df1['low'] = df.low.resample('W').min()
df1.head()
```

```python
# 汇总统计
# 周成交量
df1['volume(sum)'] = df['volume'].resample('W').sum()
# 周成交额
df1['money(sum)'] = df['money'].resample('W').sum()
df1.head()
```
