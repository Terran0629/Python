```python
from pyecharts import options as opts
from pyecharts.charts import Geo
from pyecharts.globals import ChartType, SymbolType

# 创建 Geo 实例
geo = Geo(init_opts=opts.InitOpts(theme=ThemeType.DARK))
geo.add_schema(maptype='china')
# 定义城市及其坐标
cities_coords = {
    "北京": (116.46, 39.92),
    "上海": (121.48, 31.22),
    "广州": (113.23, 23.16),
    "深圳": (114.07, 22.62)
}

# 添加城市坐标
for city, coord in cities_coords.items():
    geo.add_coordinate(city, coord[0], coord[1])

# 定义行动轨迹数据
data = [("北京", "上海"), ("上海", "广州"), ("广州", "深圳"), ("深圳", "北京")]

# 添加行动轨迹
geo.add("",data,type_=ChartType.LINES,
    effect_opts=opts.EffectOpts(
        symbol=SymbolType.ARROW, symbol_size=6, color="yellow", trail_length=0.7
    ),
    linestyle_opts=opts.LineStyleOpts(
        color="cyan", width=3, curve=0.3, opacity=0.6
    )
)

# 设置全局配置
geo.set_global_opts(
    title_opts=opts.TitleOpts(title="美化后的行动轨迹图"),
    visualmap_opts=opts.VisualMapOpts(is_piecewise=True) # 添加视觉映射配置
)

# 渲染图表
geo.render_notebook()

```
