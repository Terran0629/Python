KD指标(又称随机指标)

默认参数：（9,3,3）

组成：K线,D线,20轴,80轴


```python
'''
KD指标计算

计算“未成熟随机值”RSV
RSV = (当前收盘价 - 近9天K线最低价)/(近9天K线最高价 - 近9天K线最低价) *100

计算K值(组成快线)
K值 = 2/3*昨日K值 + 1/3*今日RSV

计算D值(组成慢线)
D值 = 2/3*昨日D值 + 1/3*今日K值

RSV计算结果会介于0 ~100之间,如果RSV结果越大,表示今日收盘价越接近过去9天内股价的高点;
反之,RSV结果越小,则代表今日收盘价越接近过去9天内股价的低点。

K值是以今日RSV与昨日K值的加权平均,由于K值反应较为灵敏,因此K值组成了快线。
D值则是再一次把加权平均过的K值与昨日D值再一次平滑,因此D值反应较不灵敏,组成了慢线。

KD指标的应用
当K值“由下往上穿过”D值,就称为黄金交叉,这个现象出现通常被视为未来一段时间股价上涨的几率较高,建议买进。
当K值“由上往下跌破”D值,就称为死亡交叉,这个现象出现通常被视为未来一段时间股价下跌的几率较高,建议卖出。
如果只看日KD,信号过于频繁,容易有骗线的状况,将资金白白花在进出成本。
所以，从不同周期看 KD,并搭配其他技术指标,才能找出真正的进场点!
不可迷信任何指标,KD指标也不例外!
'''
```
