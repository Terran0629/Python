```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

x = range(5,11)

plt.subplot(121)
plt.plot(x,x,color=sns.xkcd_rgb['green'])
plt.subplot(122)
plt.bar(x,x,color=sns.color_palette())
```
