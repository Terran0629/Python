```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'SimHei'

#三天中三部电影的票房变化
real_names=['电影1','电影2','电影3']
real_num1=[5453,7548,6543]
real_num2=[1840,4013,3421]
real_num3=[1080,1673,2342]
#设置画布大小
plt.figure(figsize=(10,8),dpi=100)
x = range(len(real_names))
width = 0.3 

x_name = [f'第{i}天' for i in range(1,4)]
plt.xticks([x+width for x in x],x_name)
plt.xlabel(xlabel='天')
plt.ylabel(ylabel='票房数')

plt.bar(x,real_num1,width=width)
plt.bar([x+width for x in x],real_num2,width=width)
plt.bar([x+(width*2) for x in x],real_num3,width=width)

```
