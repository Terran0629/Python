```python
from pyecharts.charts import Line
from pyecharts.globals import ThemeType
from pyecharts.faker import Faker
import pyecharts.options as opts

line = Line(init_opts=opts.InitOpts(theme=ThemeType.DARK))
line.add_xaxis(Faker.choose())
line.add_yaxis('商品A',Faker.values())
line.add_yaxis('商品B',Faker.values())
line.add_yaxis('商品C',Faker.values())

line.render_notebook()
```

```python
line = Line(init_opts=opts.InitOpts(theme=ThemeType.DARK))
line.add_xaxis(Faker.choose())
line.add_yaxis('商品A',Faker.values(),stack='stack')
line.add_yaxis('商品B',Faker.values(),stack='stack')
line.add_yaxis('商品C',Faker.values(),stack='stack')
line.set_global_opts(title_opts=opts.TitleOpts(title='堆叠折线图'),
                     tooltip_opts=opts.TooltipOpts(trigger='axis'))
line.render_notebook()
```

```python
line = Line(init_opts=opts.InitOpts(theme=ThemeType.DARK))
line.add_xaxis(Faker.choose())
line.add_yaxis('商品A',Faker.values(),stack='stack')
line.add_yaxis('商品B',Faker.values(),stack='stack')
line.add_yaxis('商品C',Faker.values(),stack='stack')
line.set_series_opts(areastyle_opts=opts.AreaStyleOpts(opacity=0.6))
# 去掉x轴左右两边空白
line.set_global_opts(xaxis_opts=opts.AxisOpts(boundary_gap=False))
line.render_notebook()
```
