```python
from pyecharts.charts import Pie
from pyecharts.globals import ThemeType
from pyecharts.faker import Faker
import pyecharts.options as opts

pie = Pie(init_opts=opts.InitOpts(theme=ThemeType.DARK))
pie.add('',data_pair=[list(z) for z in zip(Faker.choose(),Faker.values())])
pie.render_notebook()
```

```python
pie = Pie(init_opts=opts.InitOpts(theme=ThemeType.DARK))
pie.add('',data_pair=[list(z) for z in zip(Faker.choose(),Faker.values())],label_opts=opts.LabelOpts(formatter='{c}'))
pie.render_notebook()
```

```python
pie = Pie(init_opts=opts.InitOpts(theme=ThemeType.DARK))
pie.add('A3pyecharts/08绘制饼图.ipynb',data_pair=[list(z) for z in zip(Faker.choose(),Faker.values())],
        label_opts=opts.LabelOpts(formatter='{c}'),
        rosetype='radius')
pie.render_notebook()
```
