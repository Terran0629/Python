```python
from pyecharts.charts import Radar
from pyecharts.globals import ThemeType
from pyecharts.faker import Faker
import pyecharts.options as opts
from random import randint


rader = Radar(init_opts=opts.InitOpts(theme=ThemeType.DARK))
rader.add_schema([
    opts.RadarIndicatorItem('语文',max_=100),
    opts.RadarIndicatorItem('数学',max_=100),
    opts.RadarIndicatorItem('化学',max_=100),
    opts.RadarIndicatorItem('英语',max_=100),
    opts.RadarIndicatorItem('物理',max_=100),
])
rader.add('小明',[[97,89,68,78,67]])
rader.add('小虎',[[90,70,78,62,87]])
rader.render_notebook()
```
