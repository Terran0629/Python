```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False    # 用来正常显示负号

x = np.linspace(0,10,100)
# print(x)
sin = np.sin(x)
plt.scatter(x,sin)
plt.show()
# plt.plot(x,sin,'o')

```
