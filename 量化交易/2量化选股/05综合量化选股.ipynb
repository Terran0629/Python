```python
from jqdatasdk import *
'''
换手率(在valuation表)大于20
流动资产合计大于流动负债合计(在balance表)
经营活动现金流入小计大于经营活动现金流出小计(在cash_flow表)
净利润(在income表)大于20000元
销售毛利率(在indicator表)大于15
'''
```

```python
# 查询的表为indicator财务指标表(也可以指定其他表)
# 查询条件来源于多张表，写法“表名.字段名”
q = query(indicator).filter(
    # 换手率（在valuation表）大于20
    valuation.turnover_ratio > 20,

    # 流动资产合计大于流动负债合计（在balance表）
    balance.total_current_assets > balance.total_current_liability,

    # 经营活动现金流入小计大于经营活动现金流出小计
    cash_flow.subtotal_operate_cash_inflow > cash_flow.subtotal_operate_cash_outflow,

    # 净利润（在income表）大于20000元
    income.net_profit > 20000,

    # 销售毛利率（在indicator表）大于15
    indicator.gross_profit_margin > 15
)

df = get_fundamentals(q,'2023-06-02')
df.head()
```
