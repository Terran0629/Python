```python
import matplotlib.pyplot as plt
import seaborn as sns


data = sns.load_dataset('tips')  # 默认文件格式为csv

plt.subplot(311)
sns.violinplot(x='day',y='total_bill',data=data)
plt.subplot(312)
sns.violinplot(x='day',y='total_bill',data=data,hue='time') # hue用于给x变量进行进一步的细分分组
plt.subplot(313)
sns.violinplot(x='day',y='total_bill',data=data,hue='sex') # hue用于给x变量进行进一步的细分分组
```

```python
plt.subplot(121)
sns.violinplot(x='day',y='total_bill',data=data,hue='sex',split=True)
plt.subplot(122)
sns.violinplot(x='day',y='total_bill',data=data,hue='sex',split=True,inner=None)
```
