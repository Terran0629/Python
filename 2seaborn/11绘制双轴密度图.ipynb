```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np


#  
#,

data = np.random.multivariate_normal([0,1] ,[(1,0.5),(0.5,1)],1000) # ,[(1,0.5),(0.5,1)]
df = pd.DataFrame(data, columns=['x','y'])
plt.figure(figsize=(10,6))
plt.subplot(131)
sns.kdeplot(x='x',y='y',data=df)
plt.subplot(132)
sns.kdeplot(x='x',y='y',data=df,fill=True)
plt.subplot(133)
sns.kdeplot(x='x',y='y',data=df,fill=True,color='red')
#
```
