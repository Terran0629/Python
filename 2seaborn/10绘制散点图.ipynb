```python
import matplotlib.pyplot as plt
import seaborn as sns


data = sns.load_dataset('tips')  # 默认文件格式为csv

plt.figure(figsize=(12,10))

# 分布散点图
plt.subplot(221)
sns.scatterplot(x='total_bill',y='tip',data=data)
plt.subplot(222)
sns.scatterplot(x='total_bill',y='tip',data=data,size='size')
plt.subplot(223)
sns.scatterplot(x='total_bill',y='tip',data=data,size='size',hue='smoker')
plt.subplot(224)
sns.scatterplot(x='total_bill',y='tip',data=data,size='size',hue='smoker',style='time')
```
